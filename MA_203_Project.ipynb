**FOR DOUBLE WELL POTENTIAL**

```python
m = 0.511 * (10**6)
h = 1973
a = 10
n = 1000
d = a/n
x = np.arange(-a/2, a/2 + d, d)

def V(x):
    return (x**2 - 8)**2

A = np.zeros((n-1, n-1))

for i in range(n-1):
    xi = x[i+1]
    A[i, i] = 2 * (1 + (m*d*d / (h*h)) * V(xi))
    if i < n-2:
        A[i, i+1] = -1
        A[i+1, i] = -1

eigenvalues, eigenvectors = eigh(A)

X = eigenvectors

psi0 = np.concatenate(([0], X[:, 0], [0]))
psi1 = np.concatenate(([0], X[:, 1], [0]))
psi2 = np.concatenate(([0], X[:, 2], [0]))
psi3 = np.concatenate(([0], X[:, 3], [0]))

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, psi0, '-o', label='Ground State')
plt.plot(x, psi1, '-*', label='First Excited State')
plt.plot(x, psi2, '-+', label='Second Excited State')
plt.plot(x, psi3, '-^', label='Third Excited State')
plt.legend()
plt.xlabel('x')
plt.ylabel('Psi')
plt.title('Wave Functions')

plt.subplot(1, 2, 2)
z = np.max(psi0**2) + 0.01
plt.plot(x, psi0**2, '-', label='Ground State')
plt.plot(x, psi1**2 + z, '-', label='First Excited State')
plt.plot(x, psi2**2 + 2*z, '-', label='Second Excited State')
plt.plot(x, psi3**2 + 3*z, '-', label='Third Excited State')
plt.plot(x, V(x)/500, 'k', label='Potential Energy / 150')
plt.legend()
plt.xlabel('x')
plt.ylabel('Psi^2')
plt.title('Squared Wave Functions and Scaled Potential Energy')

plt.tight_layout()
plt.show()
```
